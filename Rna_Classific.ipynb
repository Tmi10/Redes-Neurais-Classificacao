```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('data.csv', sep=',' , encoding='iso-8859-1')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Unnamed: 32'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df = df.drop(columns=['id'])
```

```python
df.head()
```

```python
df.shape
```

```python
previsores = df.iloc[:, 1:32].values
previsores
```

```python
previsores.shape
```

```python
alvo = df.iloc[:, 0].values
alvo
```

```python
from sklearn.preprocessing import LabelEncoder

alvo = LabelEncoder().fit_transform(alvo)
```

```python
alvo
```

```python
from sklearn.preprocessing import StandardScaler

previsores_esc = StandardScaler().fit_transform(previsores)

previsores_esc
```

```python
# Base de treino 

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(previsores_esc, alvo, test_size = 0.3, random_state = 0)
```

```python
x_treino.shape
```

```python
x_teste.shape
```

```python
y_treino.shape
```

```python
y_teste.shape
```

```python
# Algoritmo Rede Neural 

from sklearn.neural_network import MLPClassifier

redes = MLPClassifier(hidden_layer_sizes=(16), activation='relu', solver='lbfgs', max_iter =200,
                              tol=0.0001, random_state = 3, verbose = True)
                               
```

```python
redes.fit(x_treino, y_treino)
```

```python
previsoes_rna = redes.predict(x_teste)
```

```python
previsoes_rna
```

```python
# Classificação e Métricas

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_rna) * 100.0))
```

```python
confusion_matrix(y_teste, previsoes_rna)
```

```python
print(classification_report(y_teste, previsoes_rna))
```

```python
# Análise treino

previsoes_treino_rna = redes.predict(x_treino)
previsoes_treino_rna
```

```python
accuracy_score(y_treino, previsoes_treino_rna)
```

```python
confusion_matrix(y_treino, previsoes_treino_rna)
```

```python
# Validação Cruzada

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

# Separando os dados em folds
kfold = KFold(n_splits = 30, shuffle=True, random_state = 5)
```

```python
# Criando o modelo
modelo_rna = MLPClassifier(hidden_layer_sizes=(16), activation='relu', solver='lbfgs', max_iter =200,
                              tol=0.0001, random_state = 3, verbose = True)
resultado_rna = cross_val_score(modelo_rna, previsores_esc, alvo, cv = kfold)
```

```python
print("Acurácia Média: %.2f%%" % (resultado_rna.mean() * 100.0))
```
